```python
import pandas as pd
import numpy as np

data = {
    'Product':['Laptop','Mobile','Tablet','Headphones','smartwatch','Camera'],
    'Price':[55000,20000,30000,2000,8000,45000],
    'Units_Sold':[25,60,35,80,50,15]
}

df = pd.DataFrame(data)
print(df)
```

```python
#Data Binding

bins1 = [0,10000,30000,60000]
labels1 = ['Low','Medium','High']

df['Price_Category'] = pd.cut(df['Price'], bins=bins1, labels=labels1)

print(df)
```

```python
#Binning units sold

bins2 = [0,30,60,100]
labels2 = ['Low Sales','Medium Sales','High Sales']

df['Items_category'] = pd.cut(df['Units_Sold'], bins = bins2, labels = labels2)

print(df)
```

```python
# datatype of columns
# data formatting

# datatype covertion

print(df.dtypes)
df['Units_Sold'] = df['Units_Sold'].astype(float)
print(df.dtypes)
print(df)
```

```python
#covert text into uppercase

df['Product'] = df['Product'].str.upper()
print(df)
```

```python
#rounding Values

df['Price'] = df['Price'].round(2)
print(df)
```

```python
#Additional opertions
#Sort Data

sorted_df = df.sort_values(by='Price')
print(sorted_df)
```

```python
df.sort_values(by='Units_Sold', inplace=True)
print(df)
```

```python
df.sort_values(by='Units_Sold',ascending=False,inplace=True)
print(df)
```

```python
#Dataset: Food Delivery Orders

import pandas as pd

import numpy as np

data = {

'Order_ID': ['01', '02', '03', '04', '05', '06', '07', '08'],

'Order_Value': [250,450, 120, 600, 300, 150,500,220],

'Delivery_Time': [30,45,25,60,35,20,50,28],

'Distance_km': [5,8,3,10,6,2,9,4]

}

data= pd.DataFrame(data)

print(data)
```

```python
data.sort_values(by='Delivery_Time', inplace=True)
print(data)
```

```python
bins3 = [0,200,400,600]
labels3= ['Low','Medium','High']

data['Order_category'] = pd.cut(data['Order_Value'], bins = bins3, labels = labels3)

print(data)
```

```python
bins4 = [0,30,45,60]
labels4 = ['Fast','Normal','Slow']

data['Delivery_speed'] = pd.cut(data['Delivery_Time'], bins=bins4, labels=labels4)

print(data)
```
